```python
def Age_group(Age):
# If the age is 45 or older, 'Old' is returned, indicating that the person is in the older age group.
    if Age >= 45:
        return 'Old'
# If the age is between 25 and 44 (25 inclusive), 'Average' is returned, indicating that the person is in that average age range.
    elif Age >= 25:
        return 'Average'
# For ages under 25, 'Young' is returned, indicating that the person is in the younger age group.
    else:
        return 'Young'
```

```python
# Applies a function, probably 'Age_group', to each value in the 'Age' column to categorize age and saves the result in a new 'Pclass' column.

data_indexed['Age_category'] = data_indexed['Age'].apply(Age_group)
```
